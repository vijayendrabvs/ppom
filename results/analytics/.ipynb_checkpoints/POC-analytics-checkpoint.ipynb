# Regression model

```python
def metrics_eval(y, predict):
    """
    Evaluation metrics.
    """
    
    mae = metrics.mean_absolute_error(y, predict)
    print("Mean absolute deviation: {0}".format(mae))
    plt.figure(figsize=(15, 7))
    plt.plot(y, label="Actual", alpha=0.5, color="blue")
    plt.plot(predict, label="Predict", alpha=0.8, color="red")
    plt.yscale("log")
    plt.legend(loc="best")
    plt.title("Actual and predicted latencies")
    plt.show()
```

## Process data

```python
y = df["latency"].values
X = df.drop(labels="latency", axis=1)
```

```python
X = pd.get_dummies(X, columns=["cpus", "mem", "stress_cpus", "stress_io", "stress_vm", "stress_vm_bytes"],
               drop_first=True)
```

## OLS regression

```python
from sklearn import linear_model
from sklearn.model_selection import cross_val_predict
from sklearn import metrics
```

```python
reg = linear_model.LinearRegression(normalize=True)
reg.fit(X, y)                                     
```

```python
predict = cross_val_predict(reg, X, y, cv=3)
```

```python
metrics_eval(y, predict)
```

## Lasso regression

```python
reg = linear_model.Lasso(alpha=100, normalize=True)
predict = cross_val_predict(reg, X, y, cv=3)       
metrics_eval(y, predict)
```

## Ridge

```python
reg = linear_model.Ridge(alpha=100, normalize=True)
predict = cross_val_predict(reg, X, y, cv=3)       
metrics_eval(y, predict)
```

## Random forest regression

```python
from sklearn import ensemble
```

```python
reg = ensemble.RandomForestRegressor(n_estimators=50, criterion="mae")
predict = cross_val_predict(reg, X, y, cv=5)       
metrics_eval(y, predict)
```

## SVM

```python
from sklearn import svm
```

```python
reg = svm.SVR(kernel="rbf")
predict = cross_val_predict(reg, X, y, cv=5)       
metrics_eval(y, predict)
```

```python
reg = svm.SVR(kernel="sigmoid")
predict = cross_val_predict(reg, X, y, cv=5)       
metrics_eval(y, predict)
```
